```python
import pandas as pd
```

```python
excel = '/Users/Ketki/Documents/MS Theses/Reference Papers and Data/Outputs/commonwords_jd_china.docx.xlsx'
```

```python
df = pd.DataFrame()
for i in [0,1,2,3,4,6,7,8]:
    malay = pd.read_excel(excel, sheet_name = i)
    malay = malay[['word','Product']]
    malay = malay.set_index('word')
    df = pd.concat([df,malay],axis=1)
df
```

```python
df = df.sum(axis = 1)
df
```
